````python
detector = ObjectDetector(model_size='n', confidence_threshold=0.5)
print(f'Model loaded: YOLOv8{detector.model_size}')
````

## Part 2: Camera Intrinsics

Pinhole camera model:
```
X = (u - cx) * Z / fx
Y = (v - cy) * Z / fy
Z = depth
```

````python
intrinsics = CameraIntrinsics(fx=525.0, fy=525.0, cx=319.5, cy=239.5, width=640, height=480)
print(f'Camera: {intrinsics.width}x{intrinsics.height}')
````

## Part 3: Detection Example

````python
# Create synthetic test image
test_image = np.random.randint(0, 255, (480, 640, 3), dtype=np.uint8)
detections = detector.detect_objects(test_image)
print(f'Detected {len(detections)} objects')
````
